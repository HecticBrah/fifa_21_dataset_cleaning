#### Cleaning and converting Height column to have the same height type

```python
# Function to convert height to centimeters
def convert_to_cm(height_str):
    if isinstance(height_str, str):
        if "'" in height_str or '"' in height_str:
            # Convert height from feet and inches to centimeters
            parts = height_str.replace('"', '').split("'")
            if len(parts) == 2:
                feet, inches = map(int, parts)
                total_inches = feet * 12 + inches
                return total_inches * 2.54
            elif len(parts) == 1:
                return int(parts[0]) * 2.54
        elif "cm" in height_str:
            # Remove 'cm' and convert to numeric
            return pd.to_numeric(height_str.replace('cm', '').strip())
    # If height contains just numbers or is a float, leave it as it is
    return height_str

# Apply the conversion function to the 'Height' column
df['Height'] = df['Height'].apply(convert_to_cm)

# Display the modified DataFrame
print(df)
```

 #### Cleaning and converting Weight column to have the same height type

```python
# Function to convert weight to kilograms
def convert_to_kg(weight_str):
    # Extract numeric part
    numeric_part = ''.join(filter(str.isdigit, weight_str))
    
    if numeric_part:
        if 'lbs' in weight_str:
            # Convert to kg
            weight_in_lbs = float(numeric_part)
            weight_in_kg = weight_in_lbs * 0.453592
            return weight_in_kg
        elif 'kg' in weight_str:
            # Remove 'kg' and convert to numeric
            return pd.to_numeric(weight_str.replace('kg', '').strip())
    # If it doesn't contain 'lbs' or 'kg', leave it as it is
    return weight_str
    
# Apply the conversion function to the 'Weight' column and save results in a new column 'Weight_kg'
df['Weight'] = df['Weight'].apply(convert_to_kg)

# Display the modified DataFrame
print(df)
```

#### Making a function to remove € symbol, if row contains k value multiply it with 1000, if it contains m value multiply it with 1000000, this function will be used in multiple columns

```python
def modify_value(value_str):
    # Remove € symbol
    value_str = value_str.replace('€', '')

    if 'k' in value_str.lower():
        return float(value_str.lower().replace('k', '')) * 1000
    elif 'm' in value_str.lower():
        return float(value_str.lower().replace('m', '')) * 1000000
    else:
        return float(value_str)
```

#### Using modify_value function on a Value column

```python
df['Value'] = df['Value'].apply(modify_value)
```
